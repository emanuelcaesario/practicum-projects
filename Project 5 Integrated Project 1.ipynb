## Data Description

Here is a description of the data used for our analysis:

- `Name`: The name of the game.
- `Platform`: The platform(s) on which the game is available.
- `Year_of_Release`: The year the game was released.

Here is the meaning of each rating based on the ESRB website:

- **E** (Everyone): Content suitable for all ages.
- **T** (Teen): Content suitable for teens aged 13 and above.
- **M** (Mature): Content suitable for adults aged 17 and above.
- **E10+** (Everyone 10 and older): Content suitable for everyone over 10 years old.
- **K-A** (Kids to Adults): Content suitable for all ages. This was the rating used before it was changed to E in 1999.
- **AO** (Adults Only): Content only suitable for adults aged 18 and above.
- **EC** (Early Childhood): Content suitable for children aged 3 and above.
- **RP** (Rating Pending): Ratings have not been issued by the ESRB yet.

#### Filling in Missing Values ​​and Changing Data Types

We will begin by examining the distribution of data in the columns with missing values to identify any patterns. This will help us determine the appropriate steps for filling in the missing values.

We have identified two columns with inaccurate data types:

1. `year_of_release` is currently stored as a float type, which causes the year information to have decimal points. To address this, we will round off the values. Since this column also contains 'NaN' values, we cannot directly convert it to an integer type. Therefore, our approach will involve filling in the missing values first and then converting the data type.

2. `user_score` is currently stored as an object type. However, the content of this column appears to be similar to the `critic_score` column, which is of type float. We have noticed that the `user_score` column contains float numbers, indicating that the object type may be present due to the presence of string data. Our plan is to first convert the data type to float and then address the missing values in this column.

By performing these processes simultaneously, we will ensure that the missing values are appropriately filled in while adjusting the data types accordingly.

```python
# Checking for missing values
dataset.isna().sum()
```

##### The `name` Column

We will examine the data in the `name` column.

```python
# Checking for NaNs in the `name` column
dataset[pd.isnull(dataset['name'])]
```

The 'NaN' value in the `name` column apparently contains missing values ​​in many of the other columns, including `genre`. We'll just delete these two columns.

```python
# Removes missing rows in the `name` column
dataset = dataset[pd.notnull(dataset['name'])]
dataset.isna().sum()
```

##### The `year_of_release` Column

To fill in the missing values in the `year_of_release` column, we will leverage the relationship between the release year and the game platform. Our approach will involve using the platform information from the `platform` column as partial references.

To achieve this, we will define a custom function that takes the platform as input and returns the median value of `year_of_release` for that specific platform. This custom function will allow us to fill in the missing values based on the median release year associated with each platform.

By applying this custom function to the missing values in the `year_of_release` column, we can effectively fill in the gaps based on the median release year for each corresponding platform.

```python
# Define a function to fill in missing values
def fill_missing_value(data, agg_column, value_column):
    grouped_values = data.groupby(agg_column)[value_column].median().reset_index()
    size = len(grouped_values)
    for i in range(size):
        group = grouped_values[agg_column][i]
        value = grouped_values[value_column][i]
        data.loc[(data[agg_column]==group) & (data[value_column].isna()), value_column] = value
    return data
```

```python
# Implement the function on the 'platform' and 'year_of_release' columns 
fill_missing_value(dataset, 'platform', 'year_of_release')
```

```python
# Checking the result
dataset['year_of_release'].isna().sum()
```

It's time to change the data type.

```python
# Change the data type
dataset['year_of_release'] = dataset['year_of_release'].astype('int')
```

##### The `critic_score` column

The `critic_score` column can be associated with the `genre` column. Therefore, we will use the same function as above to fill in this column.

```python
# Implement the function on the 'genre' and 'critic_score' columns 
fill_missing_value(dataset, 'genre', 'critic_score')
```

```python
# Checking the result
dataset['critic_score'].isna().sum()
```

##### The `user_score` column

Before implementing the method to fill in the missing values in the `user_score` column, we need to address the data type issue.

The `user_score` column contains floating-point values, but its data type is currently set as object. This discrepancy may be due to the presence of string data within the column. To further understand the contents of this column, we will examine its data.

By examining the contents of the `user_score` column, we can gain insights into the nature of the data and determine the appropriate steps for data type conversion and subsequent missing value filling.

```python
# Check the contents of the `user_score` column
dataset['user_score'].unique()
```

Apparently, the `user_score` column contains string values instead of float values due to the presence of the 'tbd' value, which stands for 'to be determined'. We will check the number of occurrences of 'tbd' in the column.

```python
# Counts the sum of each unique value
dataset['user_score'].value_counts()
```

The number of 'tbd' entries is quite significant, with 2,424 rows. Therefore, it would be best not to delete them.

Since the 'tbd' entries do not provide specific information about the values, we will replace them with 'NaN' and change the data type of this column to 'float'.

```python
# Replace tbd with NaN and change the data type
dataset['user_score'] = dataset['user_score'].replace({'tbd' : np.nan})
dataset['user_score'] = dataset['user_score'].astype('float')
```

We have replaced 'tbd' with 'NaN'. Next, we will fill in these missing values using the same method as the previous column.

```python
# Implement the function on the 'genre' and 'user_score' columns 
fill_missing_value(dataset, 'genre', 'user_score')
```

```python
# Checking the result
dataset['user_score'].isna().sum()
```

##### The `rating` column

We'll first check the unique values ​​included in this column.

```python
# Checks for unique values in the `rating` column
dataset['rating'].unique()
```

The `rating` column does not contain numeric data, so we cannot use the median to fill in the missing values. Instead, we will create a new function that uses the mode (the value that occurs most frequently) and associate it with the `genre` column.

```python
# Create a new function to fill in missing values using mode
def fill_missing_rating(data, agg_column, value_column):
    grouped_values = data.groupby(agg_column)[value_column].agg(pd.Series.mode).reset_index()
    size = len(grouped_values)
    for i in range(size):
        group = grouped_values[agg_column][i]
        value = grouped_values[value_column][i]
        data.loc[(data[agg_column]==group) & (data[value_column].isna()), value_column] = value
    return data
```

```python
# Implement the function on the genre and rating column
fill_missing_rating(dataset, 'genre', 'rating')
```

```python
# Checking the result
dataset['rating'].isna().sum()
```

Previously, we received information that the rating 'K-A' or 'Kids to Adults' is another name for 'E' (Everyone). The 'K-A' rating was used before 1999 until it was finally changed to 'E'. Therefore, we will replace 'K-A' with 'E' in the `rating` column.

```python
# Changing the value in the `rating` column
dataset['rating'] = dataset['rating'].replace({'K-A' : 'E'})
dataset['rating'].unique()
```

```python
# Recheck the general information of the data
dataset.info()
```

All missing values have been filled in, and the data types have been appropriately adjusted. We now have a total of 16,713 rows of data, which is ready for analysis.

## Hypothesis Testing

We have completed the data analysis process, obtained insightful findings, and provided recommendations for video game marketing in 2017. Now, it is time to perform hypothesis tests using statistical methods.

We have been assigned two hypotheses to test:

1. Null hypothesis (H0): The average user rating for Xbox One and PC platforms is the same.
   Alternative hypothesis (H1): The average user rating for Xbox One and PC platforms is different.

2. Null hypothesis (H0): The average user rating for the Action and Sports genres is the same.
   Alternative hypothesis (H1): The average user rating for the Action and Sports genres is different.

### Hypothesis 1: The average user rating for Xbox One and PC platforms is the same.

The null hypothesis (H0) and alternative hypothesis (H1) for the test are as follows:
- H0: The average user rating for the Xbox One and PC platforms is the same.
- H1: The average user rating for the Xbox One and PC platforms is different.

To test these hypotheses, we will use the `st.ttest_ind()` method, which compares the means of two independent samples. First, we will create separate data populations for the Xbox One and PC platforms based on the `user_rating` column. We will use the population data for the entire year to ensure an adequate sample size for the test.

Before conducting the hypothesis test, we need to check if the variances of the two populations are equal. This can be done using an appropriate statistical test.

The alpha value (significance level) for the test is set at 5%. If the p-value resulting from the test is less than 0.05, we will reject the null hypothesis (H0) and conclude that the average user ratings for the two platforms are different. Conversely, if the p-value is greater than or equal to 0.05, we will accept the null hypothesis (H0) and conclude that the average user ratings for the two platforms are the same.

```python
# Create data population for the Xbox One platform
dataset_xone = dataset.query('platform == "XOne"')
dataset_xone
```

```python
# Create data population for PC platforms
dataset_pc = dataset.query('platform == "PC"')
dataset_pc
```

**Next, we will examine the variance values ​​of these two datasets**

```python
dataset_xone['user_score'].var()
```

```python
dataset_pc['user_score'].var()
```

Since the variance values of the two datasets are not significantly different, we will assume them to be equal in the hypothesis test. Therefore, we will set the parameter 'equal_var' as true.

Now, let's proceed with the hypothesis test.

```python
alpha = 0.05
results = st.ttest_ind(dataset_xone['user_score'], dataset_pc['user_score'], equal_var=True)
print('p-value:', results.pvalue)
if (results.pvalue < alpha):
    print("We reject H0")
else:
    print("We accept H0")
```

Based on the results of the hypothesis testing, we reject the null hypothesis (H0) and accept the alternative hypothesis (H1).

Therefore, we can conclude that **the average user rating for the Xbox One and PC platforms is different.**

### Hypothesis 2: The average user ratings for the Action and Sports genres are different.

The following is the null hypothesis (H0) and the alternative hypothesis (H1) that we will test:

- H0: The average user rating for the Action and Sports genres is the same.
- H1: The average user rating for the Action and Sports genres is different.

We will create a new dataset that includes only the 'Action' and 'Sports' genres. Then, we will perform a hypothesis test based on the `user_score` column to test whether the average user rating differs between these two genres.

```python
# Create data population for the Action genre
dataset_action = dataset.query('genre == "Action"')
dataset_action
```

```python
# Create data population for the Sports genre
dataset_sports = dataset.query('genre == "Sports"')
dataset_sports
```

**Next, we will examine the variance values ​​of these two datasets**

```python
dataset_action['user_score'].var()
```

```python
dataset_sports['user_score'].var()
```

The variance values of the two datasets are also not significantly different, so we will proceed with the hypothesis test assuming equal variances.

Now, we will perform the hypothesis test to determine whether the average user rating for the Action and Sports genres is different.

```python
alpha = 0.05
results = st.ttest_ind(dataset_action['user_score'], dataset_sports['user_score'], equal_var=True)
print('p-value:', results.pvalue)
if (results.pvalue < alpha):
    print("We reject H0")
else:
    print("We accept H0")
```

Based on the results of the hypothesis testing, we have decided to accept the null hypothesis (H0). Therefore, we conclude that the average user rating for the Action and Sports genres is the same.
