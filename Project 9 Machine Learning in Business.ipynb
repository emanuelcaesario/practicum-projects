## 4 Modeling and Analysis

### 4.1 Split the Data and Train the Model for Each Region

Here are the steps we followed:
- Split the data into a training set and a validation set with a ratio of 75:25.
- Train the model and make predictions for the validation set.
- Save the predictions and actual values for the validation set.
- Calculate the average volume of predicted oil reserves and the RMSE of the model.

#### 4.1.1 Consolidate Data in a Single List

To streamline the machine training process in the function, we will combine all three datasets into a single list. Additionally, we will remove the 'id' column as it is not considered relevant for model training.

```python
# Unite data in a single list
data_all = [
    df0.drop('id', axis=1),
    df1.drop('id', axis=1),
    df2.drop('id', axis=1),
]
```

#### 4.1.2 Split the Data and Train the Model

```python
state = np.random.RandomState(12345)

# These two variables will be used for the purposes of calculating profit in the next section.
# For now it can be ignored first.  
samples_target = []
samples_predict = []

for region in range(len(data_all)):
    data = data_all[region]
    
    features = data.drop('product', axis=1)
    target = data['product']
    
    features_train, features_valid, target_train, target_valid = train_test_split(
        features, target, test_size=0.25, random_state=state)
    
    model = LinearRegression()
    model.fit(features_train, target_train)
    prediction = model.predict(features_valid)
    
    samples_target.append(target_valid.reset_index(drop=True))
    samples_predict.append(pd.Series(prediction))
    
    mean_product_target = target.mean()
    mean_product_valid = target_valid.mean()
    mean_product_predict = prediction.mean()
    model_rmse = mean_squared_error(target_valid, prediction)**0.5
    
    print("-- Region", region, "--")
    print("Mean Product Target Amount =", mean_product_target)
    print("Mean Product Valid Amount =", mean_product_valid)
    print("Mean Product Predict Amount =", mean_product_predict)
    print("Model RMSE =", model_rmse)
    print()
```

#### 4.4.1 See Profit Distribution with *Bootstrapping* Technique

We will utilize the *bootstrapping* technique to generate 1000 samples of targets and associate them with their corresponding predicted values. Following the original instructions, we will select a sample of 500 wells (stored in the `total_sample` variable). From this sample, we will calculate the profit for the top 200 wells in each region (stored in the `top_sample` variable).

To facilitate this process, we will create a function to calculate the profit of each site in the `top_sample` for each region.

```python
# Setting up the function for profit calculation
def profit_prep(target, predictions):
    top_samples_predict = predictions.sort_values(ascending=False)[:top_sample]
    top_samples_target = target[top_samples_predict.index]
    total_volume = top_samples_target.sum()
    profit = total_volume * revenue_per_product - budget
    return profit 
```

Next, we will generate a profit distribution using the *bootstrapping* technique with 1,000 samples.

```python
# Create a function for bootstrapping profit distribution
def profit_dist(target, predictions, region):
    state = np.random.RandomState(12345)
    target = target[region]
    predictions = predictions[region]
    
    profit_values = []
    
    for i in range(1000):
        predictions_sample = predictions.sample(total_sample, replace=True, random_state=state)
        target_sample = target[predictions_sample.index]
        profit_values.append(profit_prep(target_sample, predictions_sample))
    return pd.Series(profit_values)
```

```python
# Visualization of profit distribution
profit_reg0 = profit_dist(samples_target, samples_predict, 0)
profit_reg1 = profit_dist(samples_target, samples_predict, 1)
profit_reg2 = profit_dist(samples_target, samples_predict, 2)

plots = [profit_reg0, profit_reg1, profit_reg2] 
titles = ['Region 0', 'Region 1', 'Region 2'] 
for i in range(3):
    plt.hist(plots[i], bins=20)
    plt.title(titles[i])
    plt.xlabel('Profit')
    plt.show()
```

Based on this visualization, it is evident that the majority of the 200 best points in each region are profitable.

#### 4.4.2 Average Profit, 95% Confidence Interval, and Risk of Loss

In the following analysis, we will calculate the average profit for each region using the bootstrapping method, along with a 95% confidence interval. Additionally, we will assess the risk of loss associated with investing in the 200 best points in each region.

```python
for region in range(3):
    state = np.random.RandomState(12345)
    target = samples_target[region]
    predictions = samples_predict[region]
    
    profit_values = []
    
    for i in range(1000):
        predictions_sample = predictions.sample(total_sample, replace=True, random_state=state)
        target_sample = target[predictions_sample.index]
        profit_values.append(profit_prep(target_sample, predictions_sample))
    
    profit_values = pd.Series(profit_values)
    
    mean_profit = profit_values.mean()
    confidence_interval = (profit_values.quantile(0.025), profit_values.quantile(0.975))
    negative_profit_chance = (profit_values < 0).mean()
    
    print("-- Region", region, "--")
    print("Mean Profit =", mean_profit, "USD")
    print("95% Confidence Interval =", confidence_interval)
    print("Risk of Losses =", negative_profit_chance * 100, "%")
    print()
```

Based on our calculations, we have determined that Region 1 offers the highest average profit with the lowest risk of loss. However, Region 0 also falls within an acceptable risk threshold, which is below 2.5%. Moreover, Region 0 exhibits a relatively high average profit.
