## 2 Technological Process

To complete this project, we were given a description of the technological process involved in the gold extraction. This information helps us understand the structure of the data we have obtained.

### 2.1 Two Stage Purification

We were informed that the mined ore goes through an initial processing stage to obtain a coarser ore blend or feed. This feedstock is then used for the flotation process, which aims to produce a coarser concentrate. Following flotation, the particles undergo a two-stage purification process.

Here is the process:

**1. Flotation**
The gold ore mixture is introduced into a flotation tank to separate the Au concentrate from the coarser tails, which contain lower concentrations of precious metals. The stability of this process is influenced by the volatility and suboptimal physicochemical state of the flotation pulp, which is a mixture of solid and liquid particles.

**2. Purification**
The coarse Au concentrates undergo a two-stage purification process. After the refining process, we obtain a refined metal concentrate and a tail, resulting in a new level of final metal concentration.

### 2.2 Calculation of Recovery

We were tasked with simulating the process of gold recovery from gold ore. To simulate the acquisition process, we are provided with the following formula:

\begin{equation}
    Recovery = \frac{C \cdot (F - T)}{F \cdot (C - T)} \cdot 100\%
\end{equation}


With:

- C — the percentage of gold in the concentrate right after the flotation process (to calculate the coarser concentrate gain) / after refining (to calculate the final concentrate gain)
- F — the percentage of gold in the feed before the flotation process (to calculate the coarser concentrate gain) / concentrate right after the flotation process (to calculate the final concentrate gain)
- T — the percentage of gold in the coarser tail, right after the flotation process (to calculate the coarser concentrate gain) / after refining (to calculate the final concentrate gain)

To predict the coefficients, it is necessary to determine the percentage of gold in the concentrates and tails. Both the final concentrate and the coarser concentrate play an equally important role.

### 2.3 Evaluation Metrics

To address the existing problem, we need to introduce a new metric called sMAPE (symmetric Mean Absolute Percentage Error), which measures the accuracy of our predictions.

sMAPE is similar to MAE (Mean Absolute Error), but it is expressed as a percentage relative to the target value. The term "symmetric" in sMAPE refers to the fact that it considers both the target and prediction scales.

Here is the formula to calculate sMAPE:

\begin{equation}
    sMAPE = \frac{1}{N}\sum\limits_{i=1}^N\frac{\mid y_i-\hat{y_i}\mid}{\left(\mid y_i\mid + \mid \hat{y_i}\mid\right)/2} \times 100\%
\end{equation}

Meaning:
- Yi: The target value for the observation with index i in the dataset used for quality measurement.
- Ŷi: The predicted value for the observation with index i, for example, in the test dataset.
- N: The total number of observations in the dataset.
- Σ: The sum of all observations in the dataset, where i ranges from 0 to N.

### 2.4 Target Prediction

We are tasked with predicting two values:

- Coarse concentrate gain (`rougher.output.recovery`)
- Final concentrate gain (`final.output.recovery`)

The final metric consists of these two values:


\begin{equation}
    Final~sMAPE = 25\% \times sMAPE (rougher) + 75\% \times sMAPE (final)
\end{equation}

### 3.3 Checking If Gold Recovery Has Been Calculated Properly

We have been instructed to calculate the recovery for the `rougher.output.recovery` feature using the training set. To do this, we need to apply the formula provided to estimate the gold recovery rate.

The MAE, or Mean Absolute Error, is a metric used to evaluate the accuracy of forecasting models. It represents the average absolute difference between the actual values and the predicted values.

Based on the given formula for calculating the gold recovery rate:

\begin{equation}
    Recovery = \frac{C \cdot (F - T)}{F \cdot (C - T)} \cdot 100\%
\end{equation}

We can calculate the recovery rate and then compare it with the actual values in the `rougher.output.recovery` feature to compute the MAE.

Meanwhile, each component in the equation above has the following meanings:
- C represents the percentage of gold in the concentrate right after the flotation process. In the given scheme (see below), this corresponds to **point 3** and can be found in the column `rougher.output.concentrate_au`.
- F represents the percentage of gold in the feed before the flotation process. This corresponds to **point 1** and can be found in the column `rougher.input.feed_au`.
- T represents the percentage of gold in the coarser tail right after the flotation process. This corresponds to **point 2** and can be found in the column `rougher.output.tail_au`.



```python
# Prepare clean datasets with no missing values
train_mae = df_train.dropna()
```

```python
# Doing the calculations
C = train_mae['rougher.output.concentrate_au']
F = train_mae['rougher.input.feed_au']
T = train_mae['rougher.output.tail_au']

# Create a new feature of recovery calculation results
train_mae['recovery'] = ((C * (F - T)) / (F * (C - T))) * 100
```

```python
# Show the results
display(train_mae['recovery'].head())
display(train_mae['rougher.output.recovery'].head())
```

At first glance, based on the first five rows of data, the two values appear to be the same. Now, let's proceed to calculate the MAE.

```python
# Calculate the MAE
calculate_mae = mae(train_mae['rougher.output.recovery'], train_mae['recovery'])
print('MAE value is = ', calculate_mae)
```

The MAE is negative and very small, indicating that there is no significant difference between the two datasets. In other words, the `rougher.output.recovery` feature in the `df_train` dataset is accurate.

### 4.1 Changes in the Concentration of Each Metal at Each Refining Stage

We will analyze the concentration changes of three metals: gold (Au), silver (Ag), and lead (Pb) at different production stages: `rougher`, `primary_cleaner`, and `final`. We'll ignore the `secondary_cleaner` stage since its output is the same as the final output.

The parameters we'll be considering are:
- `rougher.input.feed`
- `rougher.output.concentrate`
- `primary_cleaner.output.concentrate`
- `final.output.concentrate`

To better understand the distribution at each stage, we'll define a function for visualization.

```python
# Define a function for visualization
metals = ['au', 'ag', 'pb']
stage_parameter = ['rougher.input.feed', 
                   'rougher.output.concentrate', 
                   'primary_cleaner.output.concentrate', 
                   'final.output.concentrate']
xcolors = ['red', 'orange', 'yellow', 'green']


def metal_concentrate(df):
    kwargs = dict(hist_kws={'alpha': 0.5}, kde_kws={'linewidth':2})
    for element in metals:
        plt.figure(figsize=(10,6), dpi=80)
        for features, colour in zip(stage_parameter, xcolors):
            sns.distplot(df[features+'_'+element], color=colour, label=features+'_'+element, **kwargs)
            plt.axvline(0, c="r")
        plt.title('Distribution of ' +element+ ' concentrations across purification stage')
        plt.xlabel('Concentration of '+element)
        plt.legend();
```

We will use the function to analyze the `df_full` dataset as it contains all the data and represents the complete dataset.

```python
# Implement the function
metal_concentrate(df_full)
```

**Findings:**

After analyzing the changes in the distribution of all metals (gold, silver, and lead) at each stage of the production process, we observed the following:

- The concentration of gold (Au) tends to increase from the raw feed stage to the coarser concentrate stage and further increases at the final concentrate stage.

```python
# Create a function to determine the lower whisker
def lower_whisker (data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    value = Q1 - (1.5 * IQR)
    return value
```

## 5 Developing Models

Instructions for Stage:

- Create a function to calculate the final sMAPE value.
- Train various models, evaluate them with cross-validation, and select the best model. Test the model using a test sample.
- Use the formula for evaluation metrics:

\begin{equation}
    sMAPE = \frac{1}{N}\sum\limits_{i=1}^N\frac{\mid y_i-\hat{y_i}\mid}{\left(\mid y_i\mid + \mid \hat{y_i}\mid\right)/2} \times 100\%
\end{equation}

- Predict the two values: Coarse concentrate gain (rougher.output.recovery) and Obtain final concentrate (final.output.recovery).
- Calculate the final sMAPE using the given formula:

\begin{equation}
    Final~sMAPE = 25\% \times sMAPE (rougher) + 75\% \times sMAPE (final)
\end{equation}

### 5.1 Designing a Function to Calculate sMAPE

```python
# Function to calculate sMAPE
def smape(y_true, y_pred):
    smape = (np.abs(y_true - y_pred)/((np.abs(y_true) + np.abs(y_pred))/2)).mean() * 100
    return smape 

# Function to calculate final sMAPE
def smape_final(y_true, y_pred):
    smape_out_rougher = smape(y_true[:,0], y_pred[:,0])
    smape_out_final = smape(y_true[:,1], y_pred[:,1])
    return ((0.25 * smape_out_rougher) + (0.75 * smape_out_final))

# Scoring function
smape_score = make_scorer(smape_final)
```

### 5.2 Preparing the Datasets
