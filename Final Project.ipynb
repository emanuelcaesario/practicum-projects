## 1 Introduction

### 1.1 Project Description

A telecommunications operator named Interconnect wants to predict the churn rate of their clients. If it is known that a client is planning to quit, they will be offered promotional codes and special package options. Interconnect's marketing team has collected some personal data from clients, including information about data plans selected and their contracts.

**Interconnect Services**

Interconnect provides two main types of services:

1. Landline network: Phones can be connected to multiple channels simultaneously.
2. Internet: Internet networks can be managed via telephone lines (DSL, digital subscriber line) or via fiber optic cables.

Some of the other services that Interconnect provides include:

- Internet security: antivirus software (DeviceProtection) and malicious website blocker (OnlineSecurity)
- Dedicated technical support line (TechSupport)
- Cloud storage for files and backup data (OnlineBackup)
- Streaming TV (StreamingTV) and movie directories (StreamingMovies)

Clients can choose to pay monthly or sign a contract for a 1 or 2-year subscription. They can use various payment methods and receive electronic bills after transactions.

### 1.2 Data Description

The available data consists of four files obtained from different sources:

- `contract.csv` — containing contract information
- `personal.csv` — containing client personal data
- `internet.csv` — containing information about Internet services
- `phone.csv` — containing information about phone services

Each file includes a `customerID` column with a unique code assigned to each client.

The contract information is up to date as of February 1, 2020.

### 1.3 Project Goals

This project has three main objectives:

- Analyze the provided data to identify factors that can influence the churn rate of customers.
- Develop machine learning models that can efficiently predict the likelihood of customers leaving Interconnect services.
- Generate relevant promotional recommendations to mitigate customer churn and retain their loyalty to the service.

### 1.4 Stages of Project Completion

This project will be carried out in four phases:

1. Data pre-processing: Enhance data quality to facilitate more effective analysis.
2. Exploratory data analysis: Investigate the factors influencing customer churn rate.
3. Machine learning modeling: Develop a machine learning model capable of accurately predicting churn rate.
4. Final report and recommendations: Summarize the project implementation and propose appropriate promotion strategies based on the key factors contributing to customer churn in Interconnect services.

### 2.3 Fixing the Data

#### 2.3.1 Uniform Column Names

```python
# Creates a variable containing a list of column names
old_columns = ['customerID', 
               'BeginDate', 
               'EndDate', 
               'Type', 
               'PaperlessBilling', 
               'PaymentMethod', 
               'MonthlyCharges', 
               'TotalCharges', 
               'InternetService', 
               'OnlineSecurity', 
               'OnlineBackup', 
               'DeviceProtection', 
               'TechSupport', 
               'StreamingTV', 
               'StreamingMovies', 
               'gender', 
               'SeniorCitizen', 
               'Partner', 
               'Dependents', 
               'MultipleLines'
              ]

# Rename column
new_columns = []
for column in old_columns:
       
    # Separate words with regex except ID and TV
    if not column.endswith('ID') and not column.endswith('TV'):
        column = re.sub(r'(?<!^)(?=[A-Z])', '_', column)
    
    new_columns.append(column)

# Change all column names to lowercase strings
new_columns = [column.lower() for column in new_columns]

df.columns = new_columns

# Checking the result
df.columns
```

All column names were successfully uniformed.

#### 2.3.2 Changing the Data Type

The columns with the data type that need to be changed are as follows:

- `begin_date` and `end_date` should be converted from object to datetime type.
- `total_charges` should be converted from object to float type.
- Categorical columns 'Yes' and 'No' should be converted to 1 and 0, respectively. This applies to columns such as `online_security`, `online_backup`, `device_protection`, `tech_support`, `streamingtv`, `streaming_movies`, `partner`, `dependents`, and `multiple_lines`.

**The `begin_date` and `end_date` Columns**

```python
# Checking for unique values in `begin_date`
df['begin_date'].unique()
```

```python
# Checking for unique values in `end_date`
df['end_date'].unique()
```

The date data in both columns starts from the date 01, and the hour information is not relevant. We will extract only the month and year from these columns.

The `end_date` column contains significantly fewer values. This column only indicates whether the subscriber is still active or has churned. 

We will convert the `end_date` column into a `churn` column with values 0 and 1. A value of 0 means the customer is still active, while 1 means the customer has churned. We made this decision because the `end_date` column has only 5 unique values, representing 4 different months from October 2019 to January 2020, while the `begin_date` column contains customer data starting from October 2013 to February 2020. Therefore, we will remove the `end_date` column from the dataset.

```python
# Change data type of `begin_date` column to datatime
df['begin_date'] = df['begin_date'].apply(pd.to_datetime).dt.to_period('M')
```

```python
# Checking the result
df['begin_date'].head()
```

```python
# Change column 'end_date' to 'churn'
# Value 'No' is changed to 0 (still active), while the others are 1 (already churned)
df['churn'] = df['end_date'].apply(lambda x: 0 if x == 'No' else 1)
df = df.drop('end_date', axis=1)
```

```python
# Checking the result
df['churn'].head()
```

The data and its type changed successfully.

**The `total_charges` Column**

```python
# Checking for unique values in `total_charges`
df['total_charges'].unique()
```

There doesn't appear to be any issue. We will convert it to a numeric format.

```python
# Change the data type of `total_charges` to numeric
df['total_charges'] = pd.to_numeric(df['total_charges'], errors='coerce')
```

```python
# Checking the result
df['total_charges'].head()
```

Data type successfully changed to float64.

**The categorical columns contain `Yes` and `No`**

We have several categorical columns that currently contain only 'Yes' and 'No' values. To simplify the analysis process, we will convert these categorical values to binary, with 1 representing 'Yes' and 0 representing 'No'.

The columns we will convert are: `online_security`, `online_backup`, `device_protection`, `tech_support`, `streamingtv`, `streaming_movies`, `partner`, `dependents`, and `multiple_lines`.

```python
# Check unique values for each column other than `partner`, `depends`, and `paperless_billing` 
unique_values = pd.DataFrame({
    'online_security': df['online_security'].unique(),
    'online_backup': df['online_backup'].unique(),
    'device_protection': df['device_protection'].unique(),
    'tech_support': df['tech_support'].unique(),
    'streamingtv': df['streamingtv'].unique(),
    'streaming_movies': df['streaming_movies'].unique(),
    'multiple_lines': df['multiple_lines'].unique()
})

unique_values.T
```

Previously, we observed that the `NaN` values appeared because the initial dataset, specifically the `internet` dataset, only contains data for customers who use internet services. Therefore, customers who do not use internet services (NaN) can be treated as having the same value as 'No'.

To simplify the analysis, we will convert all the `Yes` values to 1, while the `No` and `NaN` values to 0. This will make it easier to handle the categorical columns in the dataset.

```python
# Change the value in the categorical column from Yes to 1 and No/NaN to 0
# Change the data type from object to integer

# Create a list of columns whose values will be changed
cols_to_convert = ['online_security', 
                   'online_backup', 
                   'device_protection', 
                   'tech_support', 
                   'streamingtv', 
                   'streaming_movies', 
                   'partner', 
                   'dependents', 
                   'multiple_lines',
                   'paperless_billing'
]

# Change the values in all of these columns
df[cols_to_convert] = df[cols_to_convert].replace({'Yes': 1, 'No': 0, np.nan: 0})

# Change column type to int
df[cols_to_convert] = df[cols_to_convert].astype(int)

# Displays data resulting from changes
df[cols_to_convert].head()
```

```python
# Recheck the type of all data
df.dtypes
```

```python
# Checking the contents of the data
df.head()
```

All data types have been adjusted, and the columns containing the values 'Yes' and 'No' have also been changed to 1 and 0.

**Preparing Some Functions**

We will iteratively perform several analysis processes on each column. To streamline this, we will define a custom function.

```python
# Define functions to calculate data distribution and create pie charts 
def analyze_column(column):
    # Calculates value counts and percentages 
    value_counts = df[column].value_counts()
    percentages = (value_counts / len(df)) * 100

    # Create dataframe
    data = pd.DataFrame({'Value Counts': value_counts, 'Percentages': percentages.map('{:.2f}%'.format)})

    # Display dataframe
    display(data)

    # Displays a pie chart
    plt.pie(value_counts, labels=value_counts.index, autopct='%.1f%%')

    # Added title
    plt.title(f'Distribution of {column}')

    # Displays a pie chart
    plt.show()
```

```python
# Define a function to calculate the churn rate and visualize it
def analyze_churn_rate(column):
    # Calculates churn rate by column
    churn_rate = df.groupby(column)['churn'].value_counts().unstack(fill_value=0)

    # Added churn_rate column
    churn_rate['churn_rate'] = (churn_rate[1] / churn_rate.sum(axis=1) * 100).round(2).astype(str) + '%'
    
    # Create a column-based churn rate visualization
    churn_rate.plot(kind='bar', stacked=False)

    # Added axis titles and labels
    plt.title('Churn Rate by ' + column)
    plt.xlabel(column)
    plt.ylabel('Count')

    # Show legend
    plt.legend(title='Churn', loc='upper right')
    
    # Set the x label to italic
    plt.gca().set_xticklabels(churn_rate.index, rotation=45, ha='right')

    # Display plot
    plt.show()
    
    # Returns the churn_rate dataframe
    return churn_rate
```

### 5.2 Model Training

The metrics used to evaluate the performance of the machine learning models that we will design are AUC-ROC and Accuracy. To streamline the process for various machine learning models, we will create a custom function that performs the same steps.

#### 5.2.1 Create Evaluation Function

**The `calculate_metrics` function**

In this function, we won't implement hyperparameter tuning. We'll calculate the AUC-ROC score on the train dataset using the cross-validation method, while accuracy will be computed using the predictions made on the `features_train` without cross-validation. Afterward, we'll apply the model to the test dataset to calculate both AUC-ROC and Accuracy.

```python
# Created a function to calculate AUC-ROC and Accuracy scores without hyperparameter tuning
def calculate_metrics(model, features_train, target_train, features_test, target_test, model_name):
    # Perform probability predictions on the train dataset with cross-validation techniques
    proba_predictions_train = cross_val_predict(model, features_train, target_train, cv=5, method='predict_proba')
    
    # Take positive class probability column
    positive_probabilities_train = proba_predictions_train[:, 1]
    
    # Calculating AUC-ROC scores on a train dataset
    auc_roc_train = roc_auc_score(target_train, positive_probabilities_train)
    
    # Make predictions on the train dataset
    predictions_train = model.predict(features_train)
    
    # Calculating the accuracy score on the train dataset
    accuracy_train = accuracy_score(target_train, predictions_train)
    
    # Make predictions on the test dataset
    predictions_test = model.predict(features_test)
    
    # Calculates AUC-ROC and accuracy scores on a test dataset
    auc_roc_test = roc_auc_score(target_test, predictions_test)
    accuracy_test = accuracy_score(target_test, predictions_test)
    
    # Create a result dataframe
    results_df = pd.DataFrame({'Model': [model_name],
                               'AUC-ROC Train': [auc_roc_train],
                               'AUC-ROC Test': [auc_roc_test],
                               'Accuracy Train': [accuracy_train],
                               'Accuracy Test': [accuracy_test]})
    
    return results_df
```

**The `hyperparameter_tuning` function**

In this function, we'll apply hyperparameter tuning to the model to find the best hyperparameters, aiming to achieve improved AUC-ROC and Accuracy scores. After tuning the model, we'll use it to make predictions on the test dataset and calculate the AUC-ROC and Accuracy scores based on the tuned model's performance.

```python
def hyperparameter_tuning(model, param_distributions, features_train, target_train, features_test, target_test, model_name):
    # Define a RandomizedSearchCV object
    random_search = RandomizedSearchCV(model, param_distributions=param_distributions, 
                                       scoring='roc_auc', n_iter=10, cv=5, n_jobs=-1, random_state=42)
    
    # Doing fitting on the train dataset to find the best hyperparameter combination
    random_search.fit(features_train, target_train)
    
    # Get the model with the best hyperparameters
    best_model = random_search.best_estimator_
    
    # Melakukan prediksi probabilitas pada dataset train dan test menggunakan model terbaik
    proba_predictions_train = best_model.predict_proba(features_train)
    proba_predictions_test = best_model.predict_proba(features_test)
    
    # Take positive class probability column
    positive_probabilities_train = proba_predictions_train[:, 1]
    positive_probabilities_test = proba_predictions_test[:, 1]
    
    # Calculate AUC-ROC scores on train and test datasets
    auc_roc_train = roc_auc_score(target_train, positive_probabilities_train)
    auc_roc_test = roc_auc_score(target_test, positive_probabilities_test)
    
    # Make predictions on the train and test datasets using the best models
    predictions_train = best_model.predict(features_train)
    predictions_test = best_model.predict(features_test)
    
    # Calculating accuracy on train and test datasets
    accuracy_train = accuracy_score(target_train, predictions_train)
    accuracy_test = accuracy_score(target_test, predictions_test)
    
    # Display best_params_from RandomizedSearchCV results
    print("Best Params:", random_search.best_params_)
    
    # Create a result dataframe
    results_df = pd.DataFrame({'Model': [model_name],
                               'AUC-ROC Train': [auc_roc_train],
                               'AUC-ROC Test': [auc_roc_test],
                               'Accuracy Train': [accuracy_train],
                               'Accuracy Test': [accuracy_test]})
    
    return results_df
```
